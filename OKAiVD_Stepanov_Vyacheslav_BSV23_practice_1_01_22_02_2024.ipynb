# Описание задачи

 Функция <b>Соровников П.С. No2 (2022) </b>

  $f_{19}(x) = -\sinh(x)cos(\pi x) - e^{0.1x}$

  $ x ∈ [-4.2,4.2], f_{min}(\approx 4.09863) \approx -30.1927 $

# Вычисление функции

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_beautiful_function(x):
  result = -1 * np.sinh(x)*np.cos(np.pi*x) - np.power(np.e,0.1*x)
  return result;
```

```python
x_min = -4.2
x_max = 4.2

x_opt = 4.09863
y_opt = -30.1927
number_of_points = 200
x = np.linspace(x_min, x_max, number_of_points)
print("x =", x[0:3], "...",x[-4:number_of_points])
y = np.empty(number_of_points)
for i in range(number_of_points):
  y[i] = my_beautiful_function(x[i])
print("F(x) = ", y[0:3], "...", y[-4:number_of_points])
```

```python
# Строю график
plt.plot(x,y)
```

```python
# Улучшаю график
fig, ax = plt.subplots(figsize=(10,5), dpi=75)
ax.scatter(x_opt,y_opt,color = "blue", marker='X',s=100,zorder=1,label = "min")
ax.plot(x,y, lw=2,color='red',zorder=0,label = "Функция")
ax.set_xlabel('x',fontsize=14)
ax.set_xlabel('y',fontsize=14)
ax.set_title("Функция Соровникова П.С. $f_{19}$")
ax.grid()
ax.legend()
ax.set_axisbelow(True)
plt.show()
```
