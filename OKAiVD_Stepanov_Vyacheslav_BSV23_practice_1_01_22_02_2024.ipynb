 Функция <b>Соровников П.С. No2 (2022) </b>

  $f_{19}(x) = -\sinh(x)cos(\pi x) - e^{0.1x}$

  $ x ∈ [-4.2,4.2], f_{min}(\approx 4.09863) \approx -30.1927 $

# Вычисление функций

```python
# Стандартные действия при построении графика функции

# Создание графика функции в заданом диапозоне и вызов функции визуализации
def my_beautiful_func_graph(x_min, x_max, x_opt, y_opt, author):
    number_of_points = 200
    x = np.linspace(x_min, x_max, number_of_points)
    y = np.empty(number_of_points)
    for i in range(number_of_points):
        y[i] = my_beautiful_function(x[i])

    my_beautiful_func_show(x, y, x_opt, y_opt, author)

#Функция визуализации графика
def my_beautiful_func_show(x, y, x_opt, y_opt, author):
    fig, ax = plt.subplots(figsize=(10,5), dpi=75)
    ax.scatter(x_opt, y_opt, color="blue", marker='X', s=100, zorder=1, label="min")
    ax.plot(x, y, lw=2, color='red', zorder=0, label="Функция")
    ax.set_xlabel('x', fontsize=14)
    ax.set_ylabel('y', fontsize=14)
    ax.set_title("Функция " + author)
    ax.grid()
    ax.legend()
    ax.set_axisbelow(True)
    plt.show()
```

## Соровников П.С. No2 (2022)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_beautiful_function(x):
  result = -1 * np.sinh(x)*np.cos(np.pi*x) - np.power(np.e,0.1*x)
  return result;
```

```python
x_min = -4.2
x_max = 4.2

x_opt = 4.09863
y_opt = -30.1927
number_of_points = 200
x = np.linspace(x_min, x_max, number_of_points)
print("x =", x[0:3], "...",x[-4:number_of_points])
y = np.empty(number_of_points)
for i in range(number_of_points):
  y[i] = my_beautiful_function(x[i])
print("F(x) = ", y[0:3], "...", y[-4:number_of_points])
```

```python
# Строю график
plt.plot(x,y)
```

```python
# Улучшаю график
fig, ax = plt.subplots(figsize=(10,5), dpi=75)
ax.scatter(x_opt,y_opt,color = "blue", marker='X',s=100,zorder=1,label = "min")
ax.plot(x,y, lw=2,color='red',zorder=0,label = "Функция")
ax.set_xlabel('x',fontsize=14)
ax.set_xlabel('y',fontsize=14)
ax.set_title("Функция Соровникова П.С. $f_{19}$")
ax.grid()
ax.legend()
ax.set_axisbelow(True)
plt.show()
```

## Функция Горнова А.Ю. (2007)

$
f_{1}(x) = \sin(\pi x-0.5\pi) + 0.1x^2

$ x ∈ [-4, 5], f_{min}(0)=-1$

```python
def my_beautiful_function(x):
  result = np.sin(np.pi*x - 0.5*np.pi) + 0.1*x**2
  return result;
```

## Функция Растригина

$
f_{2}(x) = x^2 - 10 \cos(2\pi x)+10
$

$ x ∈ [-5, 10], f_{min}(0)=0$

```python
def my_beautiful_function(x):
  result = x**2 - 10 * np.cos(2*np.pi*x) + 10
  return result;
```

## Функция Ackley

$ 20 + e - 20e^{-0.2\left| x \right|} - e^{\cos(2\pi x)} $


```python
def my_beautiful_function(x):
  result = 20 + np.e - 20 * np.exp(-0.2*np.abs(x)) - np.exp(np.cos(2*np.pi*x))
  return result;
```

```python
my_beautiful_func_graph(-10, 30, 0, 0, "Ackley")
```

## Функция Alpine 1

$ \left| x \sin(x) + 0.1x \right| $

## Функция Trigonometric 2

$ 1 + 8 \sin^2\left(7(x-0.9)^2\right) + 6 \sin^2\left(14(x-0.9)^2\right) + (x-0.9)^2 $


```python
def my_beautiful_function(x):
  result = 1 + 8 * np.power(np.sin(7 * np.power(x-0.9, 2)), 2) + 6 * np.power(np.sin(14 * np.power(x-0.9, 2)), 2) + np.power(x-0.9, 2)
  return result;
```

## Функция Multiextremal 2

$ 1 - 0.5 \cos\left(1.5(10x - 0.3)\right) \cos\left(31.4x\right) + 0.5 *\cos\left(\sqrt{5} \cdot 10x\right) \cos\left(35x\right) $

## Функция Levy

$ \sin^2\left(\pi \left(\frac{{x-1}}{4} + 1\right)\right) + \frac{1}{16}(x-1)^2\left(\sin\left(\pi \left(\frac{{x-1}}{4} + 1\right)\right) + 1\right) $


```python
def my_beautiful_function(x):
  result = np.power(np.sin(np.pi*((x-1)/4 + 1)),2) + 1/16 * np.power(x-1,2) * (np.sin(np.pi*((x-1)/4 + 1)) +1)
  return result;
```

## Функция HappyCat

$ {\left| x^2 - 1 \right|}^{0.25} + 0.5x^2 + x + 0.5 $

```python
def my_beautiful_function(x):
  result = np.power(np.abs(np.power(x, 2) - 1),0.25) + 0.5 * np.power(x,2) + x+0.5
  return result;
```

```python
my_beautiful_func_graph(-4, 4, -1, 0, "HappyCat")
```
